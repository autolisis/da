```python
import json
import pandas as pd
import pickle

# Pre cleaned data imported from CSV to JSOn
with open('b.json') as f:
    raw = json.load(f)
```

```python
data = [[str(y) for y in x] for x in raw]
del raw
```

```python
from collections import Counter

d = Counter()

for row in data:
    for col in row:
        d[col] += 1
with open('counter-pick', 'wb') as f:
    pickle.dump(d, f)
# with open('counter-pick', 'rb') as f:
#     d = pickle.load(f)
print(d.most_common(10))
print(d['15846407'])
```

```python
txns = []
# Remove uncommon records by sorting and slicing based on d
for row in data[:40000]:
    row = sorted(row, key=lambda x: d[x], reverse=True) 
    txns.append(row[:5])
# 
# with open('txns-dump', 'wb') as f:
#     pickle.dump(txns, f)
# with open('txns-dump', 'rb') as f:
#     txns = pickle.load(f)
```

```python
from apyori import apriori
```

```python
ruleGen = apriori(txns, min_support = 0.015, 
                 min_confidence=0.5, min_lift=1, min_length=3)
rules = list(ruleGen)
print(len(rules))
```

```python
for item in rules:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + ','.join(items[1:]))

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
